```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

time.sleep(5)

top_vids = driver.find_element(By.ID,"toc-Top_videos")
top_vids.click()


rank = []
title = []
uploader = []
views = []
date = []

ranking = driver.find_elements(By.XPATH,'//table/tbody/tr/td[1]')
for i in ranking [1:31]:
    rank.append(i.text.replace('.',''))
    
name = driver.find_elements(By.XPATH,'//table/tbody/tr/td[2]')
for i in name [1:31]:
    title.append(i.text)
    
artist = driver.find_elements(By.XPATH,'//table/tbody/tr/td[3]')
for i in artist [0:30]:
    uploader.append(i.text)    

view = driver.find_elements(By.XPATH,'//table/tbody/tr/td[4]')
for i in view [0:30]:
    views.append(i.text)
    
dates = driver.find_elements(By.XPATH,'//table/tbody/tr/td[5]')
for i in dates [0:30]:
    date.append(i.text) 
```

```python
driver.get('https://github.com/')

driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[1]/div[2]/button').click()
driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button').click()
driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a').click()
```

```python
url = []

links = driver.find_elements(By.XPATH,'//a[@class="Link"]')
for i in links:
    url.append(i.get_attribute('href'))            
```

```python
description = []

desc = driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in desc:
    description.append(i.text)
```

```python
name = []
language = []
contributors = []

for i in url:
    driver.get(i)
    time.sleep(5)
    try:
        names = driver.find_element(By.XPATH,'//strong[@itemprop="name"]')
        lang = driver.find_element(By.XPATH,'//li[@class="d-inline"]')
        cont = driver.find_element(By.XPATH,'//a[contains(.,"Contributors")]')
        name.append(names.text)
        language.append(lang.text.replace('\n',''))
        contributors.append(cont.text.replace('Contributors\n',''))
    except NoSuchElementException:
        name.append('nan')
        language.append('nan')
        contributors.append('nan')        
```

```python
trending_repos = pd.DataFrame({'Name':name,'Description':description,'Language used':language,'No. of Contributors':contributors})
```

```python
driver.get('https://archive.ics.uci.edu/')

driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]').click()

time.sleep(5)

dataset = []
url = []

for page in range(0,65):
    data = driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in data:
        dataset.append(i.text)
        url.append(i.get_attribute('href'))    
    next_button = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_button.click()
    time.sleep(3)
```

```python
date =[]
detail = []

for i in url: 
    driver.get(i)
    time.sleep(5)
    try:
        dates = driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        date.append(dates.text.split('/'))
    except NoSuchElementException:
        date.append('nan')
        
for i in url: 
    driver.get(i)
    time.sleep(5)        
    details = driver.find_elements(By.XPATH,'//div[@class="col-span-4"]')
    for item in details:
        detail.append(item.text)        
```

```python
year_date = pd.DataFrame(date,columns=['day','month','year'])
```

```python
d_type = []
task = []
f_type = []
instance = []
f_no = []

for i in range(0,3894,6):
    d_type.append(detail[i].replace('Dataset Characteristics\n',''))
    
for i in range(2,3894,6):
    task.append(detail[i].replace('Associated Tasks\n',''))    

for i in range(3,3894,6):
    f_type.append(detail[i].replace('Feature Type\n','')) 
    
for i in range(4,3894,6):
    instance.append(detail[i].replace('# Instances\n',''))  
    
for i in range(5,3894,6):
    f_no.append(detail[i].replace('# Features\n','')) 
```

```python
datasets = pd.DataFrame({'Dataset Name':dataset,'Data Type':d_type,'Associated Tasks':task,'Feature Type':f_type,'No. of Instances':instance,'No. of Features':f_no,'Year Donated':year_date['year']})
```
