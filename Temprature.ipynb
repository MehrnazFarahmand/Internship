# Observations so far:



*   There is a reasonable linear relationship between both targets (max,min) and most features
*   The correlation of some features is different between the max and the min, so we will need to use a different mix of features for each model
* since the data we have is only for summer months, we cannot see much of a relation between the date and the weather
* we have outliers in percipitation data, but in the summer months it's normal to have many days with 0 percipitation, so we can leave the distribution as it is for those columns
* there is some multicolinearity between some columns which is normal given that max and min of weather attributes will be highly correlated.we might have to remove some of these columns
* given the high number of features we could also remove the columns with very low correlation from features
* since data is ordered by date, it's best to shuffle it before modeling



```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
x1 = data.drop(['Next_Tmax','Year','lon','Solar radiation'], axis=1)
y1 = data['Next_Tmax']
```

```python
x2 = data.drop(['Next_Tmin','Year','Month','LDAPS_PPT2'], axis=1)
y2 = data['Next_Tmin']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x1_scaled = pd.DataFrame(scaler.fit_transform(x1), columns= x1.columns)
x2_scaled = pd.DataFrame(scaler.fit_transform(x2), columns= x2.columns)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif1 = pd.DataFrame()
vif1['VIF1'] = [variance_inflation_factor(x1_scaled.values,i) for i in range(len(x1_scaled.columns))]
vif1['Features1'] = x1_scaled.columns
vif1
```

```python
vif2 = pd.DataFrame()
vif2['VIF2'] = [variance_inflation_factor(x2_scaled.values,i) for i in range(len(x2_scaled.columns))]
vif2['Features2'] = x2_scaled.columns
vif2
```

```python
x1_scaled.drop(['LDAPS_Tmin_lapse'], axis=1, inplace=True)
x2_scaled.drop(['LDAPS_Tmax_lapse'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor as KNN
```

# Next Day Maximum Temp Prediction

```python
MaxAccu = 0
MaxRS = 0
for i in range(1,400):
  x_train,x_test,y_train,y_test = train_test_split(x1_scaled,y1,test_size=0.30,random_state=i)
  LR = LinearRegression()
  LR.fit(x_train,y_train)
  pred = LR.predict(x_test)
  acc = r2_score(y_test,pred)
  if acc > MaxAccu:
    MaxAccu = acc
    MaxRS = i
print('MAximum Score:',MaxAccu,'\n', 'Best Random State:',MaxRS)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x1_scaled,y1,test_size=0.30,random_state=MaxRS)
```

**Linear Regression**

```python
LR = LinearRegression()
LR.fit(x_train,y_train)
LRpred = LR.predict(x_test)
LR_trainpred = LR.predict(x_train)
print('R2_score:', r2_score(y_test,LRpred))
print('R2_score for training data:', r2_score(y_train,LR_trainpred))
print('MAE:', mean_absolute_error(y_test, LRpred))
print('MSE:', mean_squared_error(y_test, LRpred))
print('RMSE:', np.sqrt(mean_absolute_error(y_test, LRpred)))
```

**Random Forest**

```python
RFR = RandomForestRegressor()
RFR.fit(x_train,y_train)
RFRpred = RFR.predict(x_test)
RFR_trainpred = RFR.predict(x_train)
print('R2_score:', r2_score(y_test,RFRpred))
print('R2_score for training data:', r2_score(y_train,RFR_trainpred))
print('MAE:', mean_absolute_error(y_test, RFRpred))
print('MSE:', mean_squared_error(y_test, RFRpred))
print('RMSE:', np.sqrt(mean_absolute_error(y_test, RFRpred)))
```

**KNN**

```python
knn = KNN()
knn.fit(x_train,y_train)
knnpred = knn.predict(x_test)
knn_trainpred = knn.predict(x_train)
print('R2_score:', r2_score(y_test,knnpred))
print('R2_score for training data:', r2_score(y_train,knn_trainpred))
print('MAE:', mean_absolute_error(y_test, knnpred))
print('MSE:', mean_squared_error(y_test, knnpred))
print('RMSE:', np.sqrt(mean_absolute_error(y_test, knnpred)))
```

**Gradient Boosting**

```python
GBR = GradientBoostingRegressor()
GBR.fit(x_train,y_train)
GBRpred = GBR.predict(x_test)
GBR_trainpred = GBR.predict(x_train)
print('R2_score:', r2_score(y_test,GBRpred))
print('R2_score for training data:', r2_score(y_train,GBR_trainpred))
print('MAE:', mean_absolute_error(y_test, GBRpred))
print('MSE:', mean_squared_error(y_test, GBRpred))
print('RMSE:', np.sqrt(mean_absolute_error(y_test, GBRpred)))
```

**Decision Tree**

```python
DT = DecisionTreeRegressor()
DT.fit(x_train,y_train)
DTpred = DT.predict(x_test)
DT_trainpred = DT.predict(x_train)
print('R2_score:', r2_score(y_test,DTpred))
print('R2_score for training data:', r2_score(y_train,DT_trainpred))
print('MAE:', mean_absolute_error(y_test, DTpred))
print('MSE:', mean_squared_error(y_test, DTpred))
print('RMSE:', np.sqrt(mean_absolute_error(y_test, DTpred)))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(LR, x1_scaled, y1)
print(score)
print(score.mean())
print('Difference Between score mean and R2score:', r2_score(y_test,LRpred)-score.mean())
```

```python
score1 = cross_val_score(RFR, x1_scaled, y1)
print(score1)
print(score1.mean())
print('Difference Between score mean and R2score:', r2_score(y_test,RFRpred)-score1.mean())
```

```python
score2 = cross_val_score(knn, x1_scaled, y1)
print(score2)
print(score2.mean())
print('Difference Between score mean and R2score:', r2_score(y_test,knnpred)-score2.mean())
```

```python
score3 = cross_val_score(GBR, x1_scaled, y1)
print(score3)
print(score3.mean())
print('Difference Between score mean and R2score:', r2_score(y_test,GBRpred)-score3.mean())
```

```python
score4 = cross_val_score(DT, x1_scaled, y1)
print(score4)
print(score4.mean())
print('Difference Between score mean and R2score:', r2_score(y_test,DTpred)-score4.mean())
```
