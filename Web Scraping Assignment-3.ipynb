# Question2

```python
brand = []
details = []
url = []

for page in range(1,3):
    brands = driver.find_elements(By.XPATH,'//h2[@class="a-size-mini s-line-clamp-1"]')
    for i in brands:
        brand.append(i.text)
    next_button= driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()  
    time.sleep(3)
    

for page in range(1,3):
    names = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in names:
        details.append(i.text)
        url.append(i.get_attribute('href'))
    next_button= driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()  
    time.sleep(3)  
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
price = []
delivery_date = []
returns = []
availability = []

for i in url:
    driver.get(i)
    time.sleep(6)
    try:
        prices = driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        cost = prices.text
        deliveries = driver.find_element(By.XPATH,'//div[@id="deliveryBlockMessage"]')
        returning = driver.find_element(By.XPATH,'//div[@id="freeReturns_feature_div"]')
        availabilities = driver.find_element(By.XPATH,'//div[@id="availability"]')
        price.append(cost.replace('\n','.'))
        delivery_date.append(deliveries.text)
        returns.append(returning.text)
        availability.append(availabilities.text)
    except NoSuchElementException:
        price.append('nan')
        delivery_date.append('nan')
        returns.append('nan')
        availability.append('nan') 
```

```python
search_results = pd.DataFrame({"Item": details, "Brand": brand[0:120], 'Price': price, 'Return/Exchange': returns, 'Delivery Date': delivery_date, 'Stock Availability': availability, 'Link': url})
```

```python
search_results
```

```python
search_results.to_csv('amazon_search.csv')
```

```python
driver.close()
```

```python
driver.get('https://www.flipkart.com/')

item = input(str('Enter the smartphone model'))
time.sleep(3)
search_field = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/div/input')
search_field.send_keys(item)
search_button = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/button').click()
time.sleep(3)

name_colour = []
brand = [] 
RAM_ROM = [] 
all_camera = []
screen_size = []
battery =[] 
price = []
url = []

names = driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')

for i in names:
    name_colour.append(i.text.split('('))
    
        
prices = driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for i in prices:
    price.append(i.text)

    
links = driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in links:
    url.append(i.get_attribute('href'))    
```

```python
for i in url:
    driver.get(i)
    time.sleep(6)
    try:
        ramrom = driver.find_element(By.XPATH,'//li[@class="_21Ahn-"][1]')
        display = driver.find_element(By.XPATH,'//li[@class="_21Ahn-"][2]')
        cameras = driver.find_element(By.XPATH,'//li[@class="_21Ahn-"][3]')
        battery_life = driver.find_element(By.XPATH,'//li[@class="_21Ahn-"][4]')
        RAM_ROM.append(ramrom.text.split('|',1))
        screen_size.append(display.text)
        all_camera.append(cameras.text.split('|'))
        battery.append(battery_life.text.replace('Battery',''))
        
    except NoSuchElementException:
        RAM_ROM.append('nan')
        screen_size.append('nan')
        all_camera.append('nan')
        battery.append('nan')                                  
```

```python
name_col = pd.DataFrame(name_colour, columns=['name','colour'])
```

```python
ram_rom = pd.DataFrame(RAM_ROM, columns = ['ram','rom'])
```

```python
cams = pd.DataFrame(all_camera, columns= ['BACK CAM','FRONT CAM'])
```

```python
all_phones = pd.DataFrame({'Name': name_col['name'], 'Price':price, 'Colour':name_col['colour'], 'RAM': ram_rom['ram'],'ROM':ram_rom['rom'], 'Cameras': cams['BACK CAM'], 'Selfie camera': cams['FRONT CAM'], 'Display': screen_size, 'Battery': battery, 'Link': url})
```

```python
driver.get('https://www.hostelworld.com/')

driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[1]/div/div/div/div[2]/input').click()
search_field = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div[1]/div/div[1]/div/div/div[2]/input')
search_field.send_keys('London')
delay = 10
WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME,"item-content")))
london = driver.find_element(By.XPATH,'//button[@class="item-content"]')
london.click()
driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[3]/div/div/button[2]').click()
driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[3]/div/div/button').click()
time.sleep(5)

name = []
price = []
distance = []
rating = []
review = []
rate = []
url = []
description=[]

names = driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for i in names[2:33]:
    name.append(i.text)

prices = driver.find_elements(By.XPATH,'//div[@class="property-accommodation-prices"]')
for i in prices:
    cost = (i.text.replace('\n',''))
    price.append(cost.split('Dorms'))
        
distances = driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in distances:
    distance.append(i.text.replace('from city centre',''))
    
ratings = driver.find_elements(By.XPATH,'//span[@class="keyword"]')
for i in ratings[2:33]:
    rating.append(i.text)    

reviews = driver.find_elements(By.XPATH,'//div[@class="review"]')
for i in reviews:
    review.append(i.text) 
    
rates = driver.find_elements(By.XPATH,'//span[@class="number"]')
for i in rates[2:33]:
    rate.append(i.text) 
    
link = driver.find_elements(By.XPATH,'//a[@rel="noreferrer noopener"]')
for i in link[2:33]:
    url.append(i.get_attribute('href')) 
    
    
try:
    for i in url:
        driver.get(i)
        desc = driver.find_element(By.XPATH,'//div[@class="description-container"][1]')
        description.append(desc.text)
except NoSuchElementException:
    description.append('nan')
```

```python
all_prices = pd.DataFrame(price,columns=['private','dorm'])
```

```python
london_hostels = pd.DataFrame({'Name':name, 'Private Price': all_prices['private'], 'Dorm Price':all_prices['dorm'], 'Rating':rating,'Score':rate,'Total Reviews': review, 'Distance From Central London': distance,'Description':description, 'Link': url})
```
