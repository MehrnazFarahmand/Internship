```python
quote = []
author = []
quote_type = []

for page in range(0,10):
    quotes = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quotes:
        quote.append(i.text)
    next_button = driver.find_element(By.XPATH,'//li[@class="next"]').click
    
for page in range(0,10):
    authors = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in authors:
        author.append(i.text)
    next_button = driver.find_element(By.XPATH,'//li[@class="next"]').click
    
    
for page in range(0,10):
    types = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in types:
        quote_type.append(i.text)
    next_button = driver.find_element(By.XPATH,'//li[@class="next"]').click
```

```python
thousand_quotes = pd.DataFrame({'Quote': quote, 'Author':author, 'Tags': quote_type})
```
